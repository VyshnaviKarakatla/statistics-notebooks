```python
#sorting data
#sorting salary in ascending order
data1=empdata.sort_values('salary',ascending=True)
data1
```

```python
myData=empdata.sort_values('salary',ascending=True)
myData
```

```python
avgdata=empdata.groupby('dept_id')['salary'].mean()
avgdata
```

```python
#salary greater than 50000 in ascending order and groupby dept_id
#apply and assign a new column called gross_sal by increasing 100 in salary
result= (myData[myData['salary']>50000]
         .sort_values('salary',ascending=True)
         .groupby('dept_id')
         .apply(lambda n: n.assign(gross_sal=n['salary']+100)))
result
```

```python
empdata['SALARY']=empdata['SALARY'].astype(float)
empdata
```

```python
empdata['GROSS_SAL']=empdata['SALARY']+(empdata['SALARY']*0.4)+(empdata['SALARY']*0.3)+(empdata['SALARY']*0.5)
empdata
```

```python
#Adding annual salary
empdata['ANNUAL SALARY']=empdata['GROSS_SAL']+12
empdata
```

```python
#Adding varibale pay on annual salary
empdata['VARIABLE PAY']=empdata['ANNUAL SALARY']*0.3
empdata
```

```python
#deleting column
empdata=empdata.drop(columns=['VARIABLE PAY'])
empdata
```
