```python
import pandas as pd
```

```python
print('Series with numbers')
ser1=pd.Series([10,20,30,40,50])
print(ser1)
```

```python
print('Series with strings')
ser2=pd.Series(['python','AWS','java'])
print(ser2)
```

```python
print("Slicing Series")
print(ser1[1:4])
print(ser1[::-1])
print(ser2.iloc[-1])
print(ser2.head())
print(ser2.tail())
print(ser2.head(2))
print(ser2.tail(2))
```

```python
print('DataFrames with numbers')
df1=pd.DataFrame({1,2,3,4,5})
print(df1)
```

```python
print('DataFrames with strings')
df2=pd.DataFrame({'id':['101','102','103','104','105'],'course':['java','AWS','python','cloud','html']})
print(df2)
print(df2['id'])
print(df2['id'][1:3])
print(df2['course'])
print(df2['course'][1:4])
```

```python
print("Slicing Series")
print(df1[1:3])
print(df1[::-1])
print(df1.iloc[-3])
print(df2.head())
print(df2.tail())
print(df2.head(2))
print(df2.tail(2))
```
