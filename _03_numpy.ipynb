```python
for row in ar1:
    print(row)
```

```python
for row in ar1:
    for rvalue in row:
        print(rvalue,end=" ")
    print("")
```

```python
for i in ar1.flatten():
    print(i,end=" ")
```

```python
for i in np.nditer(ar1,order='C'):
                   print(i,end=' ')
```

```python
for i in np.nditer(ar1,order='F'):
                   print(i,end=' ')
```

```python
for i in np.nditer(ar1,order='F',flags=['external_loop']):
                   print(i)
```

```python
ar1=np.arange(12).reshape(3,4)
ar1
```

```python
for i in np.nditer(ar1,op_flags=['readwrite']):
                   i[...]=i*i
ar1
```

```python
for i in np.nditer(ar1,op_flags=['readwrite']):
                   i[...]=i+2
ar1
```

```python
ar1=np.arange(12).reshape(3,4)
ar1
```

```python
for i in np.nditer(ar1, order='F',flags=['external_loop'],op_flags=['readwrite']):
                   i[...]=i+2
                   print(i,end=' ')
ar1
```
