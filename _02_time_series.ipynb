```python
#generating first 45 dates of 2025
dates=pd.date_range(start='2025-01-01',periods=45,freq='D')
dates
```

```python
#generating weeks
dates_week=pd.date_range(start='2025-01-01',periods=10,freq='W')
dates_week
```

```python
#generating months
dates_monthly=pd.date_range(start='2025-01-01',periods=20,freq='ME')
dates_monthly
```

```python
#generating quarter-end
dates_QE=pd.date_range(start='2025-01-01',periods=2,freq='QE')
dates_QE
```

```python
#generating year-end
dates_YE=pd.date_range(start='2025-01-01',periods=2,freq='YE')
dates_YE
```
