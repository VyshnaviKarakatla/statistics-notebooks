```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
marks=[48,36,47,36,48,49]
subjects=['english','telugu','maths','science','hindi','social']
```

```python
plt.plot(marks,subjects)
```

```python
plt.figure(figsize=(3,4)) #size length and width
plt.plot(subjects,marks)
```

```python
plt.plot(subjects,marks,color='red',linewidth=5)
```

```python
plt.plot(subjects,marks,color='red',linewidth=5,linestyle='dotted')
```

```python
plt.plot(subjects,marks,color='red',linewidth=5,linestyle='dashed')
```
