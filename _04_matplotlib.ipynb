```python
x=[1.0,1.1,1.2,1.3,1.4,1.5,1.6]
y=[15,25,17,35,22,40,23]
```

```python
plt.figure(figsize=(4,3))
plt.xlabel("Days")
plt.ylabel("Price")
plt.title("Stock of X Products")
plt.plot(x,y,color='b')
```

```python
plt.figure(figsize=(4,3))
plt.xlabel("Days")
plt.ylabel("Price")
plt.title("Stock of X Products")
plt.plot(x,y,'r+--')
```

```python
day=[1,2,3,4,5,6]
max_t=[45,47,38,45,36,38]
min_t=[30,32,30,34,23,25]
avg_t=[35,38,38,30,32,31]
```

```python
plt.figure(figsize=(5,4))
plt.xlabel("Days")
plt.ylabel("Price")
plt.title("Stocks Price")
plt.plot(day,max_t,label='Max')
plt.plot(day,min_t,label='Min')
plt.plot(day,avg_t,label='Avg')
plt.legend(loc='lower left')
plt.legend(loc='best')
plt.legend(loc=3,fontsize='small')
plt.grid()
```
