```python
data=np.random.exponential(scale=2000,size=1000)
```

```python
log_data=np.log1p(data)
```

```python
plt.subplot(1,2,1)
plt.hist(data,bins=50,edgecolor='black')
plt.title('Original skewed data')

plt.subplot(1,2,2)
plt.hist(log_data,bins=50,edgecolor='red')
plt.title('Logarithm Transformed data')
```

```python
df1=pd.DataFrame({
    'hour':np.arange(0,24),
    'usage_kwh':[1.2,2.4,1.3,6.2,1.5,3.0,3.5,4.0,9.3,7.4,8.9,4.9,4.2,3.5,9.0,4.5,5.2,4.1,5.6,6.3,7.3,1.5,6.3,5.3]
    })
df1
```

```python
df1.head()
```

```python
df1.tail()
```

```python
df1['hour_sin']=np.sin(2*np.pi*df1['hour']/24)
df1
```

```python
df1['hour_cos']=np.cos(2*np.pi*df1['hour']/24)
df1
```

```python
plt.title('Cyclical Encoding of Hours(cos and sin)')
plt.fig(figsize=(4,4))
plt.xlabel('cos(hour)')
plt.ylabel('sin(hour)')
```
