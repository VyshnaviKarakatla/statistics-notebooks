```python
#zeroth next percentile data point
df['price'].quantile(0.01)
```

```python
#25th percentile data point
df['price'].quantile(0.25)
```

```python
#50th percentile data point
df['price'].quantile(0.50)
```

```python
#75th percentile data point
df['price'].quantile(0.75)
```
